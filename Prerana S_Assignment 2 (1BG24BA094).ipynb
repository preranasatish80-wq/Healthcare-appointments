```python
df['AppointmentDate'] = pd.to_datetime(df['AppointmentDate'], errors='coerce')
```

```python
df.loc[df.sample(frac=0.03).index, 'City'] = np.nan
df.loc[df.sample(frac=0.03).index, 'Doctor'] = np.nan
```

```python
df['City'] = df['City'].fillna(df['City'].mode()[0])
df['Doctor'] = df['Doctor'].fillna(df['Doctor'].mode()[0])
```

```python
df['City'] = df['City'].str.strip().str.title()
df['Department'] = df['Department'].str.strip().str.title()
df['Doctor'] = df['Doctor'].str.strip().str.title()
df['VisitStatus'] = df['VisitStatus'].str.strip().str.title()
```

```python
df.rename(columns={
    'AppointmentDate': 'appointment_date',
    'ConsultationFee': 'consultation_fee',
    'VisitStatus': 'visit_status'
}, inplace=True)
```

```python
df = df.drop_duplicates()
```

```python
df = df[df['consultation_fee'] > 0]
```

```python
df.isna().sum()
```

```python
df['total_bill'] = df['consultation_fee']
```

```python
df['age_group'] = pd.cut(
    df['Age'],
    bins=[0,18,35,60,100],
    labels=['Child','Young Adult','Adult','Senior']
)
```

```python
df['high_fee_visit'] = df['consultation_fee'] >= 1500
```
